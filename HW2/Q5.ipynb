```python
# Shaojun Yu
# Mar 30
import pandas
import numpy as np
from xgboost import XGBClassifier
```

```python
#load data
data = pandas.read_csv('hw2_data_2.txt', sep='\t')
train = data[0:599]
test = data[600:].reset_index(drop=True)
train_X = train.drop(columns=['y'])
train_y = train['y']
test_X = test.drop(columns=['y'])
test_y = test['y']
```

```python
clf = XGBClassifier()
clf.fit(train_X, train_y)
print("Error rate in test data:", end=" ")
print("%.3f" % (1 - clf.score(test_X, test_y)))
```

### Feature Impotance

```python
feature_importances = pandas.DataFrame([train_X.columns, clf.feature_importances_])
feature_importances.transpose().sort_values(by=[1], ascending=False)
```
